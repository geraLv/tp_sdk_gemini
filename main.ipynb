### Interactuamos con la API de Gemini(consulta corta)

```python
from google import genai 
# amos la libreria genai
from dotenv import load_dotenv #importamos la libreria dotenv
import os #importamos la libreria os
load_dotenv() #cargamos las variables de entorno desde el archivo .env
API_KEY= os.getenv("API_KEY") #obtenemos la variable de entorno API_KEY
#creamos un cliente de genai con la API_KEY
client = genai.Client(api_key=API_KEY)
#generamos contenido con el modelo gemini-2.5-flash
response = client.models.generate_content(
model= "gemini-2.5-flash",
contents="¿Que es la inteligencia artificial?"
)
print(response.text)
```

### Con generate_content_stream(consulta corta)

```python
from google import genai #importamos la libreria genai
from dotenv import load_dotenv #importamos la libreria dotenv
import os

load_dotenv() #cargamos las variables de entorno desde el archivo .env
API_KEY= os.getenv("API_KEY") #obtenemos la variable de entorno API_KEY
#creamos un cliente de genai con la API_KEY
client = genai.Client(api_key=API_KEY)
#generamos contenido con el modelo gemini-2.5-flash
response = client.models.generate_content_stream(
model= "gemini-2.5-flash",
contents="¿Que es la inteligencia artificial?",
)
for chunk in response: #iteramos sobre la respuesta
    print(chunk.text, end="", flush=True) #imprimimos cada fragmento de la respuesta
```

### Interactuamos con la API de Gemini(consulta larga)

```python
from google import genai #importamos la libreria genai
from dotenv import load_dotenv #importamos la libreria dotenv
import os #importamos la libreria os
load_dotenv() #cargamos las variables de entorno desde el archivo .env
API_KEY= os.getenv("API_KEY") #obtenemos la variable de entorno API_KEY
#creamos un cliente de genai con la API_KEY
client = genai.Client(api_key=API_KEY)
#generamos contenido con el modelo gemini-2.5-flash
response = client.models.generate_content(
model= "gemini-2.5-flash",
contents="""¿Qué es la inteligencia artificial y como esta afecta a los puestos de trabajo donde se realizan tareas automatizables?"""
)
print(response.text)
```

### Con generate_content_stream(consulta larga)

```python
from google import genai #importamos la libreria genai
from dotenv import load_dotenv #importamos la libreria dotenv
import os

load_dotenv() #cargamos las variables de entorno desde el archivo .env
API_KEY= os.getenv("API_KEY") #obtenemos la variable de entorno API_KEY
#creamos un cliente de genai con la API_KEY
client = genai.Client(api_key=API_KEY)
#generamos contenido con el modelo gemini-2.5-flash
response = client.models.generate_content_stream(
model= "gemini-2.5-flash",
contents="""¿Qué es la inteligencia artificial y como esta afecta a los puestos de trabajo donde se realizan tareas automatizables?""",
)
for chunk in response: #iteramos sobre la respuesta
    print(chunk.text, end="", flush=True) #imprimimos cada fragmento de la respuesta
```

### Utilizamos modelos multimodales

```python
from google.genai import types
YOUR_IMAGE_PATH = './mate.jpg'
YOUR_IMAGE_MIME_TYPE = 'image/jpeg'
with open(YOUR_IMAGE_PATH, 'rb') as f:
    image_bytes = f.read()
for chunk in client.models.generate_content_stream(
model='gemini-2.0-flash-001',
contents=[
'¿Que efectos puede traer al cuerpo humano el consumo excesivo del objeto que vez en la imagen?',
types.Part.from_bytes(data=image_bytes,
mime_type=YOUR_IMAGE_MIME_TYPE),
],
):
    print(chunk.text, end='')
```

### Comparación 
* En consultas cortas el método generate_content() fue más rápido, con una diferencia de alrededor de 1.5 segundos con respecto al método generate_content_stream(). Pero la calidad de los textos generados era muy similar, solo cambia la forma en que se presenta.  

* En consultas más largas o elaboradas, el método más rápido fue el generate_content_stream(), con una diferencia de alrededor de 2 segundos con respecto al otro.  

* En conclusión, considero que el mejor método es el genetate_content_stream(), ya que, a pesar de ser más lento en consultas cortas, esta muestra en tiempo real como va generando el texto. Y le sumamos la eficiencia que tiene en consultas largas comparado con el método generate_content(), nos daremos cuenta de que es superior.
